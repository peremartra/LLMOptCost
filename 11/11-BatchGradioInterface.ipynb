```python
import gradio as gr
import pandas as pd
import time
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
# Load models
base_model_name = "bigscience/bloomz-560m"
pruned_model_name = "oopere/bloomz-560m-pruned"
custom_model_name = "oopere/bloomz-560m-pruned-kdi-kt1"
```

```python
#All models use the same tokenizer.
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
base_model = AutoModelForCausalLM.from_pretrained(base_model_name).to(device)
pruned_model = AutoModelForCausalLM.from_pretrained(pruned_model_name).to(device)
custom_model = AutoModelForCausalLM.from_pretrained(custom_model_name).to(device)
```

```python
# Load embedding model for cosine distance calculation
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

def calculate_cosine_distance(text1, text2):
    embedding1 = embedding_model.encode(text1)
    embedding2 = embedding_model.encode(text2)
    distance = cosine(embedding1, embedding2)
    return distance
```

```python
result_df=pd.DataFrame()
```

```python
def process_csv(file):
    global result_df  # Make result_df accessible globally

    df = pd.read_csv(file.name)
    results = []

    # Iterate over each row of the CSV
    for i, row in df.iterrows():
        prompt = row['text']  # Assuming 'text' column contains the prompts

        # Tokenize input
        inputs = tokenizer(prompt, return_tensors="pt").to(device)

        # Base model generation
        start_time_base = time.time()
        base_output = base_model.generate(**inputs, max_new_tokens=50)
        base_response = tokenizer.decode(base_output[0], skip_special_tokens=True)
        base_time = time.time() - start_time_base

        # Pruned model generation
        start_time_pruned = time.time()
        pruned_output = pruned_model.generate(**inputs, max_new_tokens=50)
        pruned_response = tokenizer.decode(pruned_output[0], skip_special_tokens=True)
        pruned_time = time.time() - start_time_pruned

        # Custom (Pruned + KD) model generation
        start_time_custom = time.time()
        custom_output = custom_model.generate(**inputs, max_new_tokens=50)
        custom_response = tokenizer.decode(custom_output[0], skip_special_tokens=True)
        custom_time = time.time() - start_time_custom

        # Calculate cosine distances
        cosine_distance_pruned = calculate_cosine_distance(base_response, pruned_response)
        cosine_distance_custom = calculate_cosine_distance(base_response, custom_response)

        # Append results
        results.append([prompt, base_response, pruned_response, custom_response, base_time, pruned_time, custom_time, cosine_distance_pruned, cosine_distance_custom])

    # Create DataFrame from results and make it globally accessible
    result_df = pd.DataFrame(results, columns=['Prompt', 'Base Response', 'Pruned Response',
                                               'Custom Response', 'Base Time', 'Pruned Time',
                                               'Custom Time', 'Cosine Distance Pruned',
                                               'Cosine Distance Custom'])

    # Calculate the mean values
    mean_base_time = result_df['Base Time'].mean()
    mean_pruned_time = result_df['Pruned Time'].mean()
    mean_custom_time = result_df['Custom Time'].mean()
    mean_cosine_distance_pruned = result_df['Cosine Distance Pruned'].mean()
    mean_cosine_distance_custom = result_df['Cosine Distance Custom'].mean()

    # Return the mean values
    return mean_base_time, mean_pruned_time, mean_custom_time, mean_cosine_distance_pruned, mean_cosine_distance_custom
```

```python
import gradio as gr

# Gradio interface
with gr.Blocks() as interface:
    with gr.Row():
        with gr.Column(scale=1):
            csv_file = gr.File(label="Upload CSV File with Prompts")

            # Place the textboxes in the same row
            with gr.Row():
                base_time_output = gr.Textbox(label="Mean Base Time")
                pruned_time_output = gr.Textbox(label="Mean Pruned Time")
                custom_time_output = gr.Textbox(label="Mean Custom Time")
                cosine_pruned_output = gr.Textbox(label="Mean Cosine Distance Pruned")
                cosine_custom_output = gr.Textbox(label="Mean Cosine Distance Custom")


    def handle_csv_upload(file):
        # Process the CSV and calculate the mean values
        mean_base_time, mean_pruned_time, mean_custom_time, mean_cosine_distance_pruned, mean_cosine_distance_custom = process_csv(file)
        return (
            f"{mean_base_time:.4f}",
            f"{mean_pruned_time:.4f}",
            f"{mean_custom_time:.4f}",
            f"{mean_cosine_distance_pruned:.4f}",
            f"{mean_cosine_distance_custom:.4f}"
        )

    # Update the outputs after CSV upload
    csv_file.change(
        fn=handle_csv_upload,
        inputs=csv_file,
        outputs=[
            base_time_output,
            pruned_time_output,
            custom_time_output,
            cosine_pruned_output,
            cosine_custom_output
        ]
    )

    # Removed the event listener for the display button
    # display_button.click(fn=display_result_df, outputs=result_display)

interface.launch()

```
